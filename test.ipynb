```python
import requests
import glob
import os
from pathlib import Path
import edinet
from edinet.xbrl_file import XBRLDir
import pandas as pd
```

```python
dataname = pd.read_excel("./1f_AccountList.xls",sheet_name='一般商工業',header=1)
a = dataname[["標準ラベル（英語）","名前空間プレフィックス","要素名"]].values
use = []
for i in a:
    try:
        use.append([i[0],i[1]+":"+i[2]].copy())
        pass
    except:
        pass
    pass
```

```python
avail_docs = pd.read_csv("AvailableData.csv")
data = avail_docs[(avail_docs["formCode"]=="043000")|(avail_docs["formCode"]=="030000") ]
Data = data[data["Available:xbrl"]==1]["docID"].values
```

```python
Names = []
for i in Data:
    #Download data
    DOC_ID = i
    print(DOC_ID)
    try:
        xbrl_path = edinet.api.document.get_xbrl(
            DOC_ID, save_dir="./data",
            expand_level="dir")
        xbrl_dir = XBRLDir("./data/" + DOC_ID)
        pass
    except:
        continue
        pass
    Data = []
    for i in use:
        data = []
        for net_sales in xbrl_dir.xbrl.find_all(i[1]):
            context_id = net_sales._element["contextRef"]
            context = xbrl_dir.xbrl.find("xbrli:context", {"id": context_id})
            start = xbrl_dir.xbrl.find("xbrli:startDate").text
            end   = xbrl_dir.xbrl.find("xbrli:endDate").text
            data.append([context.text, net_sales.text,context_id,start ,end ])
            pass
        Data.append([i[0],data.copy()].copy())
        pass
    df = pd.DataFrame([i for i in Data if len(i[1])!=0])
    a = df[0].apply(lambda x: "subsidy" in x).sum()
    Names.append([DOC_ID, a].copy())
    print(DOC_ID)
    pass
```
